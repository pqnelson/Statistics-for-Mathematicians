# Partial Summation: A Useful Tool

**Definition.**
Let $x\in\mathbb{R}$ be any real number. We define $\mathrm{int}(x)$ to be the
greatest integer smaller than $x$. So observe
\begin{align}
\mathrm{int}(3.1) &= 3\\
\mathrm{int}(\pi) &= 3\\
\mathrm{int}(-\mathrm{e}) &=-3
\end{align}
We always will have
\begin{equation}
\mathrm{int}(x)\leq x.
\end{equation}
We now may define
\begin{equation}
\mathrm{frac}(x)=x-\mathrm{int}(x)=\mbox{fractional part of $x$}
\end{equation}
and observe $0\leq\mathrm{frac}(x)<1$.

**Lemma.**
Suppose we have a sequence $\{c_{n}\}^{\infty}_{n=1}$ and for $x\geq1$
we define a function
\begin{equation}
C(x) = \sum_{n\leq x}c_n,\quad\mbox{and}\quad C(0)=0.
\end{equation}
Let $f$ be a $C^{1}$ function, then
\begin{equation}
\sum_{n\leq x}c_{n}f(n) = C(x)f(x)-\int^{x}_{1}C(t)f'(t)\,\mathrm{d} t.
\end{equation}

*Proof.*
This is a two-step proof. Step one notes
\begin{equation}
\sum_{n\leq x}C(n)\bigl(f(n+1)-f(n)\bigr) =
C(x)f(\mathrm{int}(x))-\sum_{n\leq x}c_{n}f(n)
\end{equation}
but the left hand side is precisely
\begin{equation}
\sum_{n\leq x}C(n)\bigl(f(n+1)-f(n)\bigr) =
\int^{\mathrm{int}(x)}_{1}C(t)f'(t)\,\mathrm{d} t.
\end{equation}
Thus we obtain
\begin{equation}
\int^{\mathrm{int}(x)}_{1}C(t)f'(t)\,\mathrm{d} t=C(x)f(\mathrm{int}(x))-\sum_{n\leq x}c_{n}f(n).\tag{1}
\end{equation}
That concludes the first step.

The second step notes
\begin{equation}
\int^{x}_{\mathrm{int}(x)}C(t)f'(t)\,\mathrm{d} t=C(x)f(x)-C(x)f(\mathrm{int}(x)).\tag{2}
\end{equation}
This uncontroversial statement should be seen immediately by the
fundamental theorem of calculus.

We then add Eq (1) to Eq (2)
to find
\begin{multline}
\int^{x}_{\mathrm{int}(x)}C(t)f'(t)\,\mathrm{d} t
+\int^{\mathrm{int}(x)}_{1}C(t)f'(t)\,\mathrm{d} t\\
=C(x)f(x)-C(x)f(\mathrm{int}(x))+C(x)f(\mathrm{int}(x))-\sum_{n\leq x}c_{n}f(n)
\end{multline}
which simplifies to
\begin{equation}
\int^{x}_{1}C(t)f'(t)\,\mathrm{d} t=C(x)f(x)+\sum_{n\leq x}c_{n}f(n)
\end{equation}
precisely as desired.
(End of Proof)

**Example (Harmonic Series).**
We will apply our lemma to the harmonic series. How? Well, consider
\begin{equation}
H_{x}=\sum_{n\leq x}\frac{1}{n}
\end{equation}
which we consider the sequence $c_{n}=1$ and $f(t)=1/t$. Then note
\begin{equation}
C(x)=\sum_{n\leq x}1=\mathrm{int}(x).
\end{equation}
Thus our lemma implies
\begin{equation}
H_{x} = \frac{\mathrm{int}(x)}{x}+\int^{x}_{1}\frac{\mathrm{int}(t)}{t^{2}}\,\mathrm{d}
t.
\end{equation}
We will try to simplify this.

First we should note that $x=\mathrm{int}(x)+\mathrm{frac}(x)$. Thus
\begin{align}
H_{x}&=\frac{x-\mathrm{frac}(x)}{x}+\int^{x}_{1}\frac{t-\mathrm{frac}(t)}{t^{2}}\mathrm{d}
t\\
&=\ln(x)+\left(1-\frac{\mathrm{frac}(x)}{x}\right)-\int^{x}_{1}\frac{\mathrm{frac}(t)}{t^{2}}\mathrm{d} t\\
&=\ln(x)+\left(1-\int^{\infty}_{1}\frac{\mathrm{frac}(t)}{t^{2}}\mathrm{d} t\right)
+\int^{\infty}_{x}\frac{\mathrm{frac}(t)}{t^{2}}\mathrm{d} t-\frac{\mathrm{frac}(x)}{x}
\end{align}
Let
\begin{equation}
\gamma=1-\int^{\infty}_{1}\frac{\mathrm{frac}(t)}{t^{2}}\mathrm{d} t
\end{equation}
then our expression simplifies to
\begin{equation}
H_{x} = \ln(x)+\gamma+\int^{\infty}_{x}\frac{\mathrm{frac}(t)}{t^{2}}\mathrm{d}
t-\frac{\mathrm{frac}(x)}{x}.
\end{equation}
Wonderful.

But does the integral term converge? We see
\begin{align}
\left|\int^{\infty}_{x}\frac{\mathrm{frac}(t)}{t^{2}}\mathrm{d} t\right|
&\leq\int^{\infty}_{x}\frac{|\mathrm{frac}(t)|}{t^{2}}\mathrm{d} t\\
&\leq\int^{\infty}_{x}\frac{1}{t^{2}}\mathrm{d} t = \frac{1}{x}.
\end{align}
Moreover this implies
\begin{equation}
\int^{\infty}_{x}\frac{\mathrm{frac}(t)}{t^{2}}\mathrm{d}
t-\frac{\mathrm{frac}(x)}{x}=\mathcal{O}(x^{-1}).
\end{equation}
So the expression for $H_{x}$ becomes
\begin{equation}
H_{x}=\ln(x)+\gamma+\mathcal{O}(x^{-1})
\end{equation}
but *can we do better?*

Lets try to figure out the next terms to order $\mathcal{O}(x^{-2})$. We will
consider
\begin{equation}
\int^{\infty}_{x}\frac{\mathrm{frac}(t)}{t^{2}}\mathrm{d} t-\frac{\mathrm{frac}(x)}{x}
=\int^{\infty}_{x}\frac{\mathrm{frac}(t)-\frac{1}{2}}{t^{2}}\mathrm{d}
t+\frac{1}{2}\int^{\infty}_{x}\frac{\mathrm{d} t}{t^{2}}-\frac{\mathrm{frac}(x)}{x}
\end{equation}
which simplifies to
\begin{equation}
\int^{\infty}_{x}\frac{\mathrm{frac}(t)}{t^{2}}\mathrm{d} t-\frac{\mathrm{frac}(x)}{x}
=\int^{\infty}_{x}\frac{\mathrm{frac}(t)-\frac{1}{2}}{t^{2}}\mathrm{d} t
+\frac{1}{2x}-\frac{\mathrm{frac}(x)}{x}.
\end{equation}
Now we make a claim!

**Claim.** The integral $\displaystyle{\int^{\infty}_{x}\frac{\mathrm{frac}(t)-\frac{1}{2}}{t^{2}}\mathrm{d} t=\mathcal{O}(1/x^{2})}$.

*Proof.*
We will integrate by parts, using
\begin{equation}
u=\frac{1}{t^{2}},\quad\mbox{and}\quad\mathrm{d} u=\frac{-2}{t^{3}}\mathrm{d} t
\end{equation}
and
\begin{equation}
\mathrm{d} v =\left(\frac{1}{2}-\mathrm{frac}(t)\right)\mathrm{d} t,\quad\mbox{and}\quad
v = \int^{t}_{1}\left(\frac{1}{2}-\mathrm{frac}(u)\right)\mathrm{d} u.
\end{equation}
Thus the integral becomes
\begin{multline}
\lim_{R\to\infty}\int^{R}_{x}\frac{\mathrm{frac}(t)-\frac{1}{2}}{t^{2}}\mathrm{d} t\\
=\left.\frac{1}{t^{2}}\int^{t}_{1}\left(\frac{1}{2}-\mathrm{frac}(u)\right)\mathrm{d} u\right|^{t=R}_{t=x}
+2\int^{R}_{x}\frac{1}{t^{3}}\int^{t}_{1}\left(\frac{1}{2}-\mathrm{frac}(u)\right)\mathrm{d} u
\end{multline}
We claim that
\begin{equation}
\begin{split}
v &= \int^{[t]}_{1}
\left(\frac{1}{2}-\mathrm{frac}(u)\right)\mathrm{d} u
+\int^{t}_{[t]}
\left(\frac{1}{2}-\mathrm{frac}(u)\right)\mathrm{d} u\\
&=0+\int^{t}_{[t]}
\left(\frac{1}{2}-\mathrm{frac}(u)\right)\mathrm{d} u
\end{split}
\end{equation}
How can we see this? Well, we should note for any integer $k$ that
\begin{equation}
\int^{k+1}_{k}\left(\frac{1}{2}-\mathrm{frac}(u)\right)\mathrm{d} u=0,
\end{equation}
since it's a sawtooth function. Thus
\begin{equation}
v\leq\int^{1}_{1/2}\left(\frac{1}{2}-\mathrm{frac}(u)\right)\mathrm{d}
u=\frac{1}{4}.
\end{equation}
Our integral becomes
\begin{equation}
\begin{split}
\lim_{R\to\infty}\int^{R}_{x}\frac{\mathrm{frac}(t)-\frac{1}{2}}{t^{2}}\mathrm{d} t
&=\lim_{R\to\infty}\left.\frac{1}{t^{2}}\frac{1}{4}\right|^{t=R}_{t=x}
+2\int^{R}_{x}\frac{1}{t^{3}}\frac{1}{4}\mathrm{d} t\\
&\sim\mathcal{O}(x^{-2}).
\end{split}
\end{equation}
This proves the claim.
(End of Proof)

Moreover, for any integer $n$, we see
\begin{equation}
\sum^{N}_{n=1}\frac{1}{n} = \ln(N)+\gamma+\frac{1}{2N}+\mathcal{O}(N^{-2}).
\end{equation}
This turns out to be a good *asymptotic* approximation for harmonic
numbers.
